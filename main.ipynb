```python
import sqlite3
import pandas as pd
import tkinter as tk
from tkinter import messagebox


# --- بخش دیتابیس ---
def setup_database():
    conn = sqlite3.connect("data.db")
    c = conn.cursor()
    c.execute('''
           CREATE TABLE IF NOT EXISTS hotel (
        FIND INTEGER PRIMARY KEY AUTOINCREMENT,
        FNAME TEXT NOT NULL,
        COST INTEGER NOT NULL,
        WEIGHT INTEGER
    )
    ''')
    conn.commit()
    conn.close()




# --- تابعی فشردن دکمه  ---
def submit_data():
    # گرفتن مقدار از تکست‌ها
    na = entry_name.get()
    co = entry_cost.get()
    wi = entry_weight.get()

    # بررسی خالی نبودن فیلدها
    if not na or not co :
        messagebox.showerror("Error", "لطفا تمام فیلدها را پر کنید")
        return

    try:
        # تبدیل قیمت و وزن به عدد
        co = int(co)
        wi = int(wi)
        
        # ذخیره 
        conn = sqlite3.connect("data.db")
        c = conn.cursor()
        c.execute("INSERT INTO hotel (FNAME, COST, WEIGHT) VALUES (?, ?, ?)", (na, co, wi))
        conn.commit()
        conn.close()
        
        # پیام موفقیت و پاک کردن فرم
        messagebox.showinfo("Success", "اطلاعات با موفقیت ذخیره شد")
        entry_name.delete(0, tk.END)
        entry_cost.delete(0, tk.END)
        entry_weight.delete(0, tk.END)
        
        
    except ValueError:
        messagebox.showerror("Error", "قیمت و وزن باید عدد باشند")

# --- ساخت رابط گرافیکی (GUI) ---

# 1. ساخت دیتابیس (اگر وجود ندارد)
setup_database()

# 2. ساخت پنجره اصلی
root = tk.Tk()
root.title("ثبت اطلاعات هتل")
root.geometry("300x250")

# 3. ایجاد لیبل‌ها و تکست‌باکس‌ها

# نام (Name)
tk.Label(root, text="Name:").pack(pady=5)
entry_name = tk.Entry(root)
entry_name.pack(pady=5)

# قیمت (Cost)
tk.Label(root, text="Cost:").pack(pady=5)
entry_cost = tk.Entry(root)
entry_cost.pack(pady=5)

# وزن (Weight)
tk.Label(root, text="Weight:").pack(pady=5)
entry_weight = tk.Entry(root)
entry_weight.pack(pady=5)
entry_weight.insert(0,"0")

# 4. دکمه ذخیره
btn_save = tk.Button(root, text="Save to Database", command=submit_data, bg="green", fg="white")
btn_save.pack(pady=20)

# اجرای برنامه
root.mainloop()
```
